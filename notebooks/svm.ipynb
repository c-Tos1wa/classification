```python
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
x, y = make_blobs(
    n_samples = 80, 
    centers = 2, 
    random_state = 6)
```

```python
fit_model = svm.SVC(kernel = "linear", C = 1000)
fit_model.fit(x, y)
```

```python
plt.scatter(
    x[:, 0],
    x[:, 1],
    c = y,
    s = 30,
    cmap = plt.cm.Paired
)

ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    fit_model,
    x,
    plot_method = "contour",
    colors = "k",
    levels = [-1, 0, 1],
    alpha = 0.5,
    linestyles = ["--", "-", "--"],
    ax = ax
)

ax.scatter(
    fit_model.support_vectors_[:, 0],
    fit_model.support_vectors_[:, 1],
    s = 100,
    linewidth = 1,
    facecolors = "none",
    edgecolors = "k"
)
plt.show()
```
